```python
train_data['VehicleType'].fillna(train_data['VehicleType'].mode().iloc[0], inplace=True)
test_data['VehicleType'].fillna(test_data['VehicleType'].mode().iloc[0], inplace=True)

vip_VT = train_data.loc[train_data['VehicleClass']==True, 'VehicleType'].unique()
train_missing_VC_idx = train_data['VehicleClass'].isna()
test_missing_VC_idx = test_data['VehicleClass'].isna()
train_data.loc[train_missing_VC_idx,'VehicleClass']=  train_data.loc[train_missing_VC_idx, 'VehicleType'].apply(lambda x: True if x in vip_VT else False)
test_data.loc[test_missing_VC_idx,'VehicleClass']=  test_data.loc[test_missing_VC_idx, 'VehicleType'].apply(lambda x: True if x in vip_VT else False)

drop_columns = ['HashPassportNumber_p', 'CancelTime', 'UserID', 'HashEmail', 'BuyerMobile', 'NationalCode', 'ReserveStatus', 'BillID', 'TicketID', 'VehicleType']
train_data.drop(columns=drop_columns, inplace=True)
test_data.drop(columns=drop_columns, inplace=True)
```

```python
train_data['DepartureTime'] = pd.to_datetime(train_data['DepartureTime'])
train_data['Created'] = pd.to_datetime(train_data['Created'])
train_data['day_to_departure'] = train_data['DepartureTime'] - train_data['Created']
train_data['day_to_departure'] = train_data['day_to_departure'].dt.days

test_data['DepartureTime'] = pd.to_datetime(test_data['DepartureTime'])
test_data['Created'] = pd.to_datetime(test_data['Created'])
test_data['day_to_departure'] = test_data['DepartureTime'] - test_data['Created']
test_data['day_to_departure'] = test_data['day_to_departure'].dt.days
```

```python
train_data['DiscountPercent'] = train_data["CouponDiscount"]/train_data["Price"]*100
test_data['DiscountPercent'] = test_data["CouponDiscount"]/test_data["Price"]*100

train_data.drop(columns=['CouponDiscount'], inplace=True)
test_data.drop(columns=['CouponDiscount'], inplace=True)

# حذف ردیف هایی که DiscountPercent مقدار بین 0 و 100 نیست.
train_data = train_data[(train_data['DiscountPercent'] >= 0) & (train_data['DiscountPercent'] <= 100)]
```

```python
train_data['Male'] = train_data['Male'].map({True:1, False:0})
```

```python
from sklearn.preprocessing import OneHotEncoder, LabelEncoder

object_columns = ['From', 'To', 'VehicleClass', 'Vehicle']
test_train_diff = []
for col in object_columns:
    test_train_diff.append([col ,len( set(train_data[col].unique()) - set(test_data[col].unique()) )])


train_test_diff = []
for col in object_columns:
    train_test_diff.append([col ,len( set(test_data[col].unique()) - set(train_data[col].unique()) )])
print(train_test_diff,'\n', test_train_diff)

all_From_cities = np.union1d(train_data['From'].unique() , test_data['From'].unique())
all_To_cities = np.union1d(train_data['To'].unique() , test_data['To'].unique())

L_encoder = LabelEncoder()
L_encoder.fit(all_From_cities)
train_data['From'] = L_encoder.transform(train_data['From'])
test_data['From'] = L_encoder.transform(test_data['From'])
L_encoder.fit(all_To_cities)
train_data['To'] = L_encoder.transform(train_data['To'])
test_data['To'] = L_encoder.transform(test_data['To'])

OH_encoder = OneHotEncoder(handle_unknown='ignore', sparse_output=False)
OH_encoder.fit(train_data[['VehicleClass', 'Vehicle']])
train_OH_encoded = OH_encoder.transform(train_data[['VehicleClass', 'Vehicle']])
test_OH_encoded = OH_encoder.transform(test_data[['VehicleClass', 'Vehicle']])

encoded_cols = OH_encoder.get_feature_names_out(['VehicleClass', 'Vehicle'])

train_encoded_df = pd.DataFrame(train_OH_encoded, columns=encoded_cols, index=train_data.index)
test_encoded_df = pd.DataFrame(test_OH_encoded, columns=encoded_cols, index=test_data.index)

train_data = train_encoded_df.combine_first(train_data)
train_data.drop(columns=['VehicleClass', 'Vehicle'], inplace=True)
test_data = test_encoded_df.combine_first(test_data)
test_data.drop(columns=['VehicleClass', 'Vehicle'], inplace=True)
```

```python
def get_season(month):
    if month in [12, 1, 2]:
        return '1'
    elif month in [3, 4, 5]:
        return '2'
    elif month in [6, 7, 8]:
        return '3'
    else:
        return '4'

train_data['DepartureSeason'] = train_data['DepartureTime'].dt.month.map(get_season)
test_data['DepartureSeason'] = test_data['DepartureTime'].dt.month.map(get_season)

train_data.drop(columns=['DepartureTime', 'Created'], inplace=True)
test_data.drop(columns=['DepartureTime', 'Created'], inplace=True)
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
scaler.fit(train_data[['Price']])
train_data['Price'] = scaler.transform(train_data[['Price']])
test_data['Price'] = scaler.transform(test_data[['Price']])

scaler.fit(train_data[['DiscountPercent']])
train_data['DiscountPercent'] = scaler.transform(train_data[['DiscountPercent']])
test_data['DiscountPercent'] = scaler.transform(test_data[['DiscountPercent']])
```

```python
from sklearn.model_selection import train_test_split
X_train, X_val, y_train, y_val = train_test_split(train_data.drop(columns=['TripReason']), train_data['TripReason'], train_size=.8)
```
